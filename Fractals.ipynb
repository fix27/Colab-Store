```
#https://habr.com/ru/company/piter/blog/496538/
#!sudo apt-get install xvfb
#!pip install PyVirtualDisplay
#!pip install pyscreenshot pillow
#!pip install xvfbwrapper
!pip3 install ColabTurtle
from ColabTurtle.Turtle import *
#import os
#from pyvirtualdisplay import Display
#from easyprocess import EasyProcess
#from pyvirtualdisplay.sma
#if 'DISPLAY' not in os.environ:
#        print("  Bringing up a new display, because os.environ does not have one.")
        #display = Display(visible=False, size=(1920, 1200))  # TODO is this cheating?? Huge resolution.
        #display.start()
        #disp = Display().start()
        #print("  Started pyvirtualdisplay with backend '%s', screen '%s'." % (disp.backend, disp.screen))
#else:
#        print("  Environ DISPLAY is '%s', no need to use pyvirtualdisplay." % os.environ['DISPLAY'])

def create_l_system(iters, axiom, rules):
    start_string = axiom
    if iters == 0:
        return axiom
    end_string = ""
    for _ in range(iters):
        end_string = "".join(rules[i] if i in rules else i for i in start_string)
        start_string = end_string

    return end_string


def draw_l_system( instructions, angle, distance):
    for cmd in instructions:
        if cmd == 'F':
            forward(distance)
        elif cmd == '+':
            right(angle)
        elif cmd == '-':
            left(angle)


def mainshow(iterations, axiom, rules, angle, length=8, size=2, y_offset=0,
        x_offset=0, offset_angle=0, width=450, height=450):

    inst = create_l_system(iterations, axiom, rules)

   # t = ColabTurtle.Turtle()
   # wn = turtle.Screen()
   # wn.setup(width, height)

    penup()
    backward(-x_offset)
    left(90)
    backward(-y_offset)
    left(offset_angle)
    pendown()
    speed(10)
   # pensize(size)
    draw_l_system(inst, angle, length)
    hideturtle()
    #wn.exitonclick()


axiom = "FX"
rules = {"X":"X+YF+", "Y":"-FX-Y"}
iterations = 8 # TOP: 16 angle = 90

angle = 60
WINDOW_SIZE = (1000, 800)
initializeTurtle(10,WINDOW_SIZE)
mainshow(iterations, axiom, rules, angle)

# 'SmartDisplay' instead of 'Display'
# It has 'waitgrab()' method.
# It has more dependencies than Display.
#with SmartDisplay() as disp:
 #   with EasyProcess(["xmessage", "hello"]):
        # wait until something is displayed on the virtual display (polling method)
        # and then take a fullscreen screenshot
      #  # and then crop it. Background is black.
      #  img = disp.waitgrab()
#img.save("xmessage.png")
#disp.stop()
```
