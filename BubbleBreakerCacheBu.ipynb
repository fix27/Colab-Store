```
#import random
import datetime
from queue import Queue
from functools import lru_cache

#(4)0,6p188;(2)0,8p176;(5)0,14p164;(1)0,9p152;(5)0,9p146;(4)2,13p74;(3)0,10p72;
#188

# simple_bubble_board = (
#                'rgrgpbyyypggrbr',
#                'gygrprrybbrrgrr',
#                'rgpbpggrbypyryr',
#                )


simple_bubble_board_s = 'rgrgpbyyypggrbrgygrprrybbrrgrrrgpbpggrbypyryr'
simple_bubble_board =  'rgrgpbyyypggrgygrprrybbrrrrgpbpggrbypyr'
                
conversion_c2i = {
    'p': 1,
    'b': 2,
    'g': 3,
    'y': 4,
    'r': 5,
    }

size_x = 3
size_y = 13

class cba_queue:

    def __init__(self, row_idx_init, col_idx_init):
        self.row_idx_v = row_idx_init
        self.col_idx_v = col_idx_init


class tans:

    def __init__(self, s, p):
        self.s = s
        self.p = p


@lru_cache(maxsize=65535)
def calc(sa, wx, wy):
    sh=0
    a = []
    for i in range(wx):
        a2 = []
        for j in range(wy):
            a2.append(int(sa[i * wy + j]))
            if sa[i * wy + j]=='0' : sh+=1
        a.append(a2)
    if sh<6: print(a)    
    v = []
    for i in range(wx):
        va2 = []
        for j in range(wy):
            va2.append(-1)
        v.append(va2)
    vn = []
    n = -1
    rez = tans('', 0)
    for i in range(wx):
      for j in range(wy):
          color = a[i][j]
          if color != 0:
              if v[i][j] == -1:
                  n += 1
                  vna2 = []
                  vna2.append(0)
                  vna2.append(i)
                  vna2.append(j)
                  vn.append(vna2)
                  fifo_cba = Queue()
                  fifo_cba.put(cba_queue(i, j))
                  while not fifo_cba.empty():
                    cba_q = fifo_cba.get()
                    ii = cba_q.row_idx_v
                    jj = cba_q.col_idx_v
                    if v[ii][jj] == -1:
                      v[ii][jj] = n
                      vn[n][0] += 1
                      if ii > 0    and color==a[ii - 1][jj] and v[ii-1][jj]==-1:
                            fifo_cba.put(cba_queue(ii - 1, jj))
                      if jj > 0    and color==a[ii][jj - 1] and v[ii][jj-1]==-1:
                            fifo_cba.put(cba_queue(ii, jj - 1))
                      if ii < wx-1 and color==a[ii + 1][jj] and v[ii+1][jj]==-1:
                            fifo_cba.put(cba_queue(ii + 1, jj))
                      if jj < wy-1 and color==a[ii][jj + 1] and v[ii][jj+1]==-1:
                            fifo_cba.put(cba_queue(ii, jj + 1))
                  if vn[n][0]<2 :
                    v[i][j] = -1
                    vn.pop()
                    n -= 1
          else:  v[i][j] = -1
    if sh<6: print(vn)
    if sh<6: print(v)
    if n >= 0 :       
     for k in range(n+1):  # =1 to n do
        tn = vn[k][0]
        if tn > 1:
            tmp = []
            for i in range(wx):
                tmpa2 = []
                for j in range(wy):
                    tmpa2.append(0)
                tmp.append(tmpa2)
            c = 0
            for i in range(wx - 1, -1, -1):  # = 11 downto 1 do
                r = True
                m = 0
                for j in range(wy - 1, -1, -1):  # =12 downto 1 do
                    if a[i][j] > 0: 
                        if v[i][j] == k:
                            m += 1
                        else:
                            tmp[i + c][j + m] = a[i][j]
                            r = False
                if r:
                    c += 1

            r = False
            while r:
             r = False
             for i in range(wx):  
                for j in range(wy-1):
                    if tmp[i][j] > 0 and tmp[i][j+1]==0 :
                         tmp[i][j+1] = tmp[i][j]
                         tmp[i][j] = 0
                         r=True            

            sc = ''
            for i in range(wx):
                for j in range(wy):
                    sc += str(tmp[i][j])
            ans = calc(sc,wx,wy)
            
            if ans.p + (tn * (tn - 1)) > rez.p:
                rez.p = ans.p + (tn * (tn - 1))
                mp = str(vn[k][1]) + ',' + str(vn[k][2]) + 'p' + str(rez.p) 
                rez.s = '(' + str(a[vn[k][1]][vn[k][2]]) + ')' +mp+ ';' +ans.s
                #if rez.p >= 170:
                #    print ( ',' + str(rez.p))
                 #   print (rez.s)
                #    print (vn)
    return rez
# random.seed()
xprint = []
for i in range(size_y):
    xprint.append(i)
print (xprint)
a = []
for i in range(size_x):
    a2 = []
    for j in range(size_y):
    # a2.insert(0,(conversion_c2i[simple_bubble_board[i*(wy)+j]]))
        a2.append(conversion_c2i[simple_bubble_board[i*size_y + j]])
    print(a2)
  # a.insert(0,  a2)
    a.append(a2)

# a = []
# for i in range(wx) :
#  a2 = []
#  for j in range(wy) :
#     #a[i][j]=random.randint(1,6)
#     a2.append(random.randint(1,5))
#  print(a2)
#  a.append(a2)

s = ''
for i in range(size_x):
    for j in range(size_y):
        s += str(a[i][j])

tb = datetime.datetime.now()
ans = calc(s,size_x,size_y)
te = datetime.datetime.now()

print (ans.s)
print (ans.p)

print ('time')
print (te)
delta = te - tb
print (delta.seconds)
```
